```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python
# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
# School Summary: School Name, Type and Budget
per_school_summary_df = school_data_complete_df[["school_name","type","budget"]]
per_school_summary_df = per_school_summary_df.drop_duplicates()
per_school_summary_df = per_school_summary_df.sort_values("school_name")
per_school_summary_df = per_school_summary_df.set_index("school_name")
per_school_summary_df
```

```python
# School Summary: Total students per school
total_students = school_data_complete_df.groupby(["school_name"]).size()
total_students
```

```python
# School Summary: Update dataframe with total students
per_school_summary_df["Total Students"] = total_students
per_school_summary_df = per_school_summary_df[["type", "Total Students", "budget"]]
per_school_summary_df = per_school_summary_df.rename(columns={"type" : "School Type",
                                                              "budget" : "Total School Budget"})
per_school_summary_df
```

```python
# School Summary: Per student budget 
per_student_budget = per_school_summary_df["Total School Budget"] / per_school_summary_df["Total Students"]
per_student_budget
```

```python
# School Summary: Average maths score
average_maths_score = school_data_complete_df.groupby(["school_name"])["maths_score"].mean()
average_maths_score
```

```python
# School Summary: Average reading score
average_reading_score = school_data_complete_df.groupby(["school_name"])["reading_score"].mean()
average_reading_score
```

```python
# School Summary: % passing maths 
students_passing_maths = school_data_complete_df.loc[(school_data_complete_df["maths_score"] >= 50)].groupby(["school_name"]).size()
students_passing_maths = students_passing_maths / per_school_summary_df["Total Students"] * 100
students_passing_maths
```

```python
# School Summary: % passing reading
students_passing_reading = school_data_complete_df.loc[(school_data_complete_df["reading_score"] >= 50)].groupby(["school_name"]).size()
students_passing_reading = students_passing_reading / per_school_summary_df["Total Students"] * 100
students_passing_reading
```

```python
# School Summary: % passing overall
students_passing_overall = school_data_complete_df.loc[(school_data_complete_df["maths_score"] >= 50) & (school_data_complete_df["reading_score"] >= 50)].groupby(["school_name"]).size()
students_passing_overall = students_passing_overall / per_school_summary_df["Total Students"] * 100
students_passing_overall
```

```python
# School Summary: combine dataframe
per_school_summary_df["Per Student Budget"] = per_student_budget
per_school_summary_df["Average Maths Score"] = average_maths_score
per_school_summary_df["Average Reading Score"] = average_reading_score
per_school_summary_df["% Passing Maths"] = students_passing_maths
per_school_summary_df["% Passing Reading"] = students_passing_reading
per_school_summary_df["% Overall Passing"] = students_passing_overall
per_school_summary_df
```

```python
# Scores by school spending: bins
spending_bins = [0, 585, 630, 645, 680]
labels_spend = ["<$585", "$585-630", "$630-645", "$645-680"]

per_school_summary_df["Spending Ranges (Per Student)"] = pd.cut(per_school_summary_df["Per Student Budget"], spending_bins, labels=labels_spend, include_lowest=True)
per_school_summary_df
```

```python
# Scores by school spending: Use code provided to calclate the following
spending_maths_scores = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Maths Score"]
spending_reading_scores = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_maths = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Maths"]
spending_passing_reading = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"] 
```

```python
# Scores by school spending: Create new dataframe
spending_summary_df = per_school_summary_df[["Spending Ranges (Per Student)"]]
spending_summary_df = spending_summary_df.drop_duplicates()
spending_summary_df = spending_summary_df.sort_values("Spending Ranges (Per Student)")
spending_summary_df = spending_summary_df.set_index("Spending Ranges (Per Student)")

spending_summary_df["Average Maths Score"] = spending_maths_scores
spending_summary_df["Average Reading Score"] = spending_reading_scores
spending_summary_df["% Passing Maths"] = spending_passing_maths
spending_summary_df["% Passing Reading"] = spending_passing_reading
spending_summary_df["% Overall Passing"] = overall_passing_spending
spending_summary_df.round(2)
```
