```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python
# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
# Maths scores by year: create series
year_9 = school_data_complete_df.loc[(school_data_complete_df["year"] == 9)].groupby(["school_name"]).mean()
year_9_maths = year_9["maths_score"]

year_10 = school_data_complete_df.loc[(school_data_complete_df["year"] == 10)].groupby(["school_name"]).mean()
year_10_maths = year_10["maths_score"]

year_11 = school_data_complete_df.loc[(school_data_complete_df["year"] == 11)].groupby(["school_name"]).mean()
year_11_maths = year_11["maths_score"]

year_12 = school_data_complete_df.loc[(school_data_complete_df["year"] == 12)].groupby(["school_name"]).mean()
year_12_maths = year_12["maths_score"]
```

```python
# Maths scores by year: create dataframe
math_year_summary_df = school_data_complete_df[["school_name"]]
math_year_summary_df = math_year_summary_df.drop_duplicates()
math_year_summary_df = math_year_summary_df.sort_values("school_name")
math_year_summary_df = math_year_summary_df.set_index("school_name")

math_year_summary_df["Year 9"] = year_9_maths
math_year_summary_df["Year 10"] = year_10_maths
math_year_summary_df["Year 11"] = year_11_maths
math_year_summary_df["Year 12"] = year_12_maths
math_year_summary_df
```

```python
# Reading scores by year: create series
year_9_reading = year_9["reading_score"]
year_10_reading = year_10["reading_score"]
year_11_reading = year_11["reading_score"]
year_12_reading = year_12["reading_score"]
```

```python
# Reading scores by year: create dataframe
reading_year_summary_df = school_data_complete_df[["school_name"]]
reading_year_summary_df = reading_year_summary_df.drop_duplicates()
reading_year_summary_df = reading_year_summary_df.sort_values("school_name")
reading_year_summary_df = reading_year_summary_df.set_index("school_name")

reading_year_summary_df["Year 9"] = year_9_reading
reading_year_summary_df["Year 10"] = year_10_reading
reading_year_summary_df["Year 11"] = year_11_reading
reading_year_summary_df["Year 12"] = year_12_reading
reading_year_summary_df
```
